```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
step_size = np.array([0.2, 1, 5, 25, 125])
df_cf_results = [(80.4, 78.6), (81, 79.2), (83.1, 81.8), (87.5, 86.3), (85.9, 80.8)]
```

```python
fig, ax = plt.subplots()
df = np.array([res[0] for res in df_cf_results])
cf = np.array([res[1] for res in df_cf_results])
ax.plot(step_size, df, marker='o', label='D-F')
ax.plot(step_size, cf, marker= 'o', label='C-F')
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.set_xlabel('step size')
ax.set_ylabel('f1')
plt.annotate(' 87.5', xy=(25, 87.5))
plt.annotate(' 86.3', xy=(25, 86.3))
# logfmt = matplotlib.ticker.LogFormatterExponent(base=5, labelOnlyBase=True)
# ax.xaxis.set_major_formatter(logfmt)
ax.set_xscale('log', base=5)
ax.legend()
# ax.set_xticks(step_size)
```
